```python
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable
from qiskit_ibm_runtime import EstimatorV2 as Estimator

import matplotlib.pyplot as plt

from qaoa.core.optimizer_strategies import NoOptimizerStrategy


graph_edges = list(graphs.edge_list())
graph_edge_weights = graphs.edges


graph_edges = np.array(graphs.edge_list())
graph_edge_weights = np.array(graphs.edges())
n_nodes = 6
def maxcut_value(bitstring):
    return sum(
        w for (u, v), w in zip(graph_edges, graph_edge_weights) if bitstring[u] != bitstring[v]
    )



quantum_no_ws = QAOArunner(
    graph=graphs,
    backend_mode="statevector",
    param_initialization="uniform",
    qaoa_variant="vanilla",
    optimizer="COBYLA",
    depth=1,
    problem_type="maxcut",
    warm_start=False,
    amount_shots=1000,
    max_tol=1e-8,
    lagrangian_multiplier=2,
)
quantum_no_ws.build_circuit()

quantum_ws = QAOArunner(
    graph=graphs,
    backend_mode="statevector",
    param_initialization="uniform",
    qaoa_variant="vanilla",
    optimizer="COBYLA",
    depth=1,
    problem_type="maxcut",
    warm_start=True,
    hamming_dist = 0,
    amount_shots=1000,
    max_tol=1e-8,
    lagrangian_multiplier=2
)
quantum_ws.build_circuit()

gamma_vals = np.linspace(0, 2 * np.pi, 50)
beta_vals = np.linspace(0, np.pi, 50)
Gamma, Beta = np.meshgrid(gamma_vals, beta_vals)
Z_no = np.zeros_like(Gamma)
Z_ws = np.zeros_like(Gamma)

estimator = Estimator(mode=quantum_no_ws.backend)
estimator.options.default_shots = 5000

no_opt_strategy = NoOptimizerStrategy(
    mode="statevector",
    backend=quantum_no_ws.backend,
    shots=5000,
)
quantum_no_ws.optimizer = no_opt_strategy
quantum_ws.optimizer = no_opt_strategy
for i in range(Gamma.shape[0]):
    for j in range(Gamma.shape[1]):
        params = np.array([Gamma[i, j], Beta[i, j]])
        Z_no[i, j] = no_opt_strategy.evaluate(params, quantum_no_ws.circuit, quantum_no_ws.cost_hamiltonian)

        Z_ws[i, j] =  no_opt_strategy.evaluate(params, quantum_ws.circuit, quantum_ws.cost_hamiltonian)

fig, ax = plt.subplots(2, 2, figsize=(16, 12))
sm = ScalarMappable(
    cmap="viridis",
    norm=Normalize(vmin=graph_edge_weights.min(), vmax=graph_edge_weights.max()),
)

try:
    edge_colors = graph_edge_weights.tolist()

    try:
        mpl_draw(
            graphs,
            ax=ax[0, 0],
            edge_color=edge_colors,
            edge_cmap=plt.get_cmap('magma'),
            edge_vmin=min(edge_colors),
            edge_vmax=max(edge_colors),
            node_color="lightgray",
            node_size=300,
            with_labels=True,
        )
    except TypeError:
        mpl_draw(graphs, ax=ax[0, 0])
except TypeError:
    mpl_draw(graphs, ax=ax[0, 0])
opt_bitstring = quantum_ws.classical_solution
ax[0, 0].set_title("Graph with edge weights color-coded")
fig.colorbar(sm, ax=ax[0, 0], label="Edge weight")

node_colors = ["tab:orange" if bit else "tab:blue" for bit in opt_bitstring]
try:
    mpl_draw(
        graphs,
        ax=ax[0, 1],
        node_color=node_colors,
        edge_color="gray",
        node_size=300,
        with_labels=True,
    )
except TypeError:
    mpl_draw(graphs, ax=ax[0, 1])
ax[0, 1].set_title(
    f"Graph with optimal cut bitstring: {''.join(str(bit) for bit in opt_bitstring)}"
)

im0 = ax[1, 0].imshow(
    Z_no,
    origin="lower",
    aspect="auto",
    extent=(gamma_vals[0], gamma_vals[-1], beta_vals[0], beta_vals[-1]),
    cmap="coolwarm",
)
ax[1, 0].set_title("Cost landscape without warm start")
ax[1, 0].set_xlabel("Gamma")
ax[1, 0].set_ylabel("Beta")
fig.colorbar(im0, ax=ax[1, 0])

im1 = ax[1, 1].imshow(
    Z_ws,
    origin="lower",
    aspect="auto",
    extent=(gamma_vals[0], gamma_vals[-1], beta_vals[0], beta_vals[-1]),
    cmap="coolwarm",
)
ax[1, 1].set_title("Cost landscape with warm start")
ax[1, 1].set_xlabel("Gamma")
ax[1, 1].set_ylabel("Beta")
fig.colorbar(im1, ax=ax[1, 1])

plt.tight_layout()
plt.show()
```

```python
from src.qaoa.models.MaxCutProblem import MaxCutProblem
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
problem = MaxCutProblem()
graphs = problem.get_erdos_renyi_graphs_paper1()
import matplotlib.pyplot as plt


graph = graphs[0]
probabilities=[]
runners = []
doping_string = None
previous_string = [0]*len(graph)
for i in range(10):
    quantum = QAOArunner(graph=graph, 
                        backend_mode = 'noisy_sampling', #statevector, noisy_sampling
                        param_initialization= 'interpolation',
                        qaoa_variant='vanilla',    
                        optimizer='COBYLA',
                        depth =1,
                        problem_type='maxcut',
                        warm_start=False,
                        maxiter = 1000,
                        max_tol = 1e-2,#lower shots and tolerance for faster runs
                        lagrangian_multiplier=2,
                        doping_string=doping_string
                        )

    quantum.build_circuit()
    print(f"\n{'='*60}")
    print(f"Iteration {i+1}/10")
    print(f"{'='*60}")
    print(f"Cost Hamiltonian: {quantum.cost_hamiltonian}")
    quantum.run()
    print(f"Solution: {quantum.solution}")
    print(f"Quantum objective value: {quantum.evaluate_solution()}")
    print(f"Classical objective value: {quantum.classical_objective_value}")
    print(f"Final expectation value: {quantum.final_expectation_value}")
    probabilities.append(quantum.get_probabilities())
    print('sum probs', np.sum([value for key, value in (quantum.get_probabilities().items())]))
    print(quantum.get_probabilities())
    print(f"{'='*60}\n")
    doping_string = [x^y for x,y in zip(quantum.solution,previous_string)]
    previous_string = quantum.solution
    print('new doping string', doping_string)
    runners.append(quantum)


fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

for iteration, prob_dict in enumerate(probabilities):
    for value, density in prob_dict.items():
        dx, dy = 0.08, 0.6
        bar = ax.bar3d(
            value - dx / 2,
            iteration - dy / 2,
            0.0,
            dx,
            dy,
            float(density),
            shade=True,
            alpha=0.9,
        )
        bar.set_array(np.array([density]))
        bar.set_cmap("viridis")
        bar.set_clim(0.0, 1.0)

ax.set_xlabel('Bitstring Value')
ax.set_ylabel('Iteration')
ax.set_zlabel('Density')
ax.set_title('Probability Distribution over Iterations')
plt.colorbar(ax.collections[0], ax=ax, label='Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

width_scale = 3.0      # controls how wide the green brush is
min_prob = 0.0         # optionally set e.g. 1e-4 to prune tiny lines

fig, ax = plt.subplots(figsize=(12, 6))

best_pts = []
zero_pts = []

for it, runner in enumerate(runners):
    ars, ps = runner.get_ar_samples()

    # optional: prune very tiny probabilities
    m = ps > min_prob
    ars, ps = ars[m], ps[m]

    # sort so it looks nice
    order = np.argsort(ars)
    ars, ps = ars[order], ps[order]

    # draw right-facing line segments
    for ar, p in zip(ars, ps):
        ax.hlines(ar, it, it + width_scale * p, color="tab:green", lw=2, alpha=0.9)

    # Best AR (choose one definition)
    best_ar = ars.max()                       # “best present”
    # best_ar = ars[np.argmax(ps)]           # or “most likely”
    best_pts.append((it, best_ar))

    # |0...0> AR
    zero_bits = [0] * runner.num_qubits
    zero_val = runner.solver.evaluate_bitstring(zero_bits)
    zero_ar = zero_val / runner.classical_objective_value if runner.classical_objective_value != 0 else 0.0
    zero_pts.append((it, zero_ar))

best_pts = np.array(best_pts)
zero_pts = np.array(zero_pts)

ax.scatter(best_pts[:, 0], best_pts[:, 1], s=60, color="cyan", label="Best AR", zorder=5)
ax.scatter(zero_pts[:, 0], zero_pts[:, 1], s=60, color="purple", label=r"$|0\ldots 0\rangle$ AR", zorder=5)

ax.set_xlabel("NDAR Iteration")
ax.set_ylabel(r"Approximation Ratio $(E/E_{GS})$")
ax.set_xlim(-0.5, len(runners) - 0.5)
ax.set_ylim(0.0, 1.02)

ax.axhline(0.9, ls="--", color="black", lw=1)
ax.legend(loc="upper center", ncol=2, frameon=True)

plt.tight_layout()
plt.show()
```
